```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import astropy.units as u

haumeadf = pd.read_csv('HaumeaObsDF.csv')
```

```python
lat1 = haumeadf['Lat-Prim']
lon1 = haumeadf['Lon-Prim']

dLat2 = haumeadf['deltaLat_Hiaka']
dLon2 = haumeadf['deltaLon_Hiaka']

dLat3 = haumeadf['deltaLat_Namaka']
dLon3 = haumeadf['deltaLon_Namaka']

lat2 = lat1+dLat2
lon2 = dLon2/np.cos(lat1*u.degree)+lon1

lat3 = lat1+dLat3
lon3 = dLon3/np.cos(lat1*u.degree)+lon1

markers = ["o" , "," , "^"]
colors = ['r','y','b']
time = haumeadf['time']

for i in range(lat1.size):
    m1 = markers[0]
    x1 = lat1[i]
    y1 = lon1[i]
    c1 = colors[0] 
    m2 = markers[1]
    x2 = lat2[i]
    y2 = lon2[i]
    c2 = colors[1] 
    m3 = markers[2]
    x3 = lat3[i]
    y3 = lon3[i]
    c3 = colors[2] 
    plt.scatter(x1,y1,marker=m1, color=c1, label = 'Haumea')
    plt.scatter(x2,y2,marker=m2, color=c2, label = 'Hiaka')
    plt.scatter(x3,y3,marker=m3, color=c3, label = 'Namaka')
    bottom, top = 185, 202
    left,right = 25.9,29
    plt.ylim(bottom, top)
    plt.xlim(left, right)
    reltime = time[i]-time[0]
    title = 'time: '+str(reltime) + " days from inital date"
    plt.title(title)
    plt.legend()
    plt.show()
#plt.show()
```
